```python
# Mapping college tier data

college_tier_mapping = {}
for tier, college in college_data.items():
     for college in college.dropna():
        college_tier_mapping[college] = tier
    
```

```python
# Mapping city type data

city_mapping = {}
for city_type, cities in cities_data.items():
    for city in cities.dropna():
        city_mapping[city] = city_type
```

```python
# Adding mapped college tier data and city type data as two new column in employee_data 

employee_data['College Tier'] = employee_data['College'].map(college_tier_mapping)
employee_data['City Type'] = employee_data['City'].map(city_mapping)
```

```python
employee_data.head()
```

```python
# As we have added college tier and city type we will remove the college and city column from employee_data

employee_data = employee_data.drop(['College', 'City'], axis = 1) 
```

```python
# Getting dummy data to covert categorical data into numerical data

employee_data = pd.get_dummies(employee_data)
employee_data
```

```python
# Removing 'Role_Executive' and'City Type_non-metro cities'

employee_data = employee_data.drop(['Role_Executive', 'City Type_non-metro cities'], axis = 1)
```

```python
# Adding 'CTC' column at the end

column_order = [col for col in employee_data.columns if col != 'CTC']

column_order.append('CTC')

# Reordering the columns in the DataFrame

employee_data = employee_data[column_order]
```

```python
# Splitting data into training and test data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)

# Creating an instance of linear Regression Model
Linear_Regression_Model = LinearRegression()

# Fitting the model on the training data
Linear_Regression_Model.fit(X_train, y_train)
```

```python
# Getting Intercept and coefficient value

print('Coefficients:', Linear_Regression_Model.coef_)
print('Intercept:', Linear_Regression_Model.intercept_)
```

```python
# Predict on the train and test data

y_test_predict = Linear_Regression_Model.predict(X_test)
y_train_predict = Linear_Regression_Model.predict(X_train)
```

```python
# Shape of training and test data

print('X_train data:', X_train.shape)
print('X_test data:', X_test.shape)
print('y_train data:', y_train.shape)
print('y_test data:', y_test.shape)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

# Calculate R2 score for train and test data
r2_train = r2_score(y_train, y_train_predict)
r2_test = r2_score(y_test, y_test_predict)
mse_lr = mean_squared_error(y_test, y_test_predict)

print('R2 Score for Train Data:', r2_train)
print('R2 Score for Test Data:', r2_test)
print('Mean Squared Error:', mse_lr)
```
